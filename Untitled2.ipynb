```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/1730285881-Airbnb_Open_Data (1).csv')

# Display basic info
df.info()

# Display first 5 rows
df.head()
```

```python
# Unique property types (room types)
df['room type'].unique()
```

```python
# Count listings per neighborhood group
listings_per_group = df['neighbourhood group'].value_counts()

print(listings_per_group)

# Visualize
plt.figure(figsize=(8,5))
sns.barplot(x=listings_per_group.index, y=listings_per_group.values, palette='viridis')
plt.title("Number of Listings per Neighborhood Group")
plt.xlabel("Neighborhood Group")
plt.ylabel("Count of Listings")
plt.show()
```

```python
avg_price = df.groupby('neighbourhood group')['price'].mean().sort_values(ascending=False)
print(avg_price)

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x=avg_price.index, y=avg_price.values, palette='coolwarm')
plt.title("Average Price per Neighborhood Group")
plt.xlabel("Neighborhood Group")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Construction year', y='price', data=df, alpha=0.5)
plt.title("Relationship between Construction Year and Price")
plt.xlabel("Construction Year")
plt.ylabel("Price")
plt.show()

# Calculate correlation
corr = df['Construction year'].corr(df['price'])
print("Correlation between construction year and price:", corr)
```

```python
top_hosts = df.groupby(['host name'])['calculated host listings count'].max().sort_values(ascending=False).head(10)
print(top_hosts)
```

```python
verified_reviews = df.groupby('host_identity_verified')['review rate number'].mean()
print(verified_reviews)

plt.figure(figsize=(6,4))
sns.barplot(x=verified_reviews.index, y=verified_reviews.values, palette='Set2')
plt.title("Average Review Rating vs Host Identity Verification")
plt.xlabel("Host Identity Verified")
plt.ylabel("Average Review Rating")
plt.show()
```

```python
corr_price_service = df['price'].corr(df['service fee'])
print("Correlation between Price and Service Fee:", corr_price_service)

plt.figure(figsize=(8,6))
sns.scatterplot(x='price', y='service fee', data=df, alpha=0.5)
plt.title("Correlation between Price and Service Fee")
plt.xlabel("Price")
plt.ylabel("Service Fee")
plt.show()
```

```python
avg_review = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().unstack()
print(avg_review)

plt.figure(figsize=(10,6))
sns.heatmap(avg_review, annot=True, cmap='YlGnBu')
plt.title("Average Review Rating by Neighborhood Group and Room Type")
plt.show()
```

```python
avg_review = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean().unstack()
print(avg_review)

plt.figure(figsize=(10,6))
sns.heatmap(avg_review, annot=True, cmap='YlGnBu')
plt.title("Average Review Rating by Neighborhood Group and Room Type")
plt.show()
```

```python
corr_availability = df['calculated host listings count'].corr(df['availability 365'])
print("Correlation between Host Listing Count and Availability:", corr_availability)

plt.figure(figsize=(8,6))
sns.scatterplot(x='calculated host listings count', y='availability 365', data=df, alpha=0.5)
plt.title("Host Listings Count vs Availability (365 days)")
plt.xlabel("Calculated Host Listings Count")
plt.ylabel("Availability 365")
plt.show()
```
